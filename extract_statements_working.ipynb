```python
from os import environ
import sec_API as sec
import json

edgar = sec.Edgar(environ.get('sec-user-agent'))
company = edgar.get_compnay_by_ticker('AMD')
```

```python
gaap = company.us_gaap
```

```python
filings = {}
for value in gaap.values():
    key = value['label']
    unit = list(value['units'].keys())[0]
    
    for file in value['units'][unit]:
        name = f"{file.get('form')}_{file.get('fp')}_{file.get('fy')}"
        filings.setdefault(name, {})[key] = {
            'value': file.get('val'), 
            'unit': unit
        }
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(filings)
df.head(15)
```
